# Collisions between particles with different masses
We searched on google, and we got the equation:
$$\textbf{v}_1' = \textbf{v}_1 - \frac{2\textbf{m}_2}{\textbf{M}}((\textbf{v}_1 -\textbf{v}_2) \cdot \hat{\textbf{r}})\hat{\textbf{r}}$$

$$\textbf{v}_2' = \textbf{v}_2 + \frac{2\textbf{m}_1}{\textbf{M}}((\textbf{v}_1 -\textbf{v}_2) \cdot \hat{\textbf{r}})\hat{\textbf{r}}$$

$$\textbf{M} = \textbf{m}_1 + \textbf{m}_2$$

where $\hat{\textbf{r}}$ is a unit vector in the direction of the vector joining the centres of the particles.

# Mean Squared Displacement-time
For this part, we have met a little problem: by this methode of programming, we must determine the time of recording the MSD of the ensemble(100 systems for average).So we use a loop to record the average MSDs at different sample time.We think it's not efficace actually 
because we didn't record the MSDs during the running of the systems (it's difficult).We have searched the theory of the diffusion of brownian
particles.For a brownian particle,
$$\textbf{MSD} = 2\textbf{Dt}$$
$\textbf{D}$ is the diffusion coefficient.

# Difficulies of the collisions between the big particle and small particles
To achieve the collisions between the big particle and small particles, we have to solve the equation:
$${\Delta\textbf{x}}^{2} + {\Delta\textbf{y}}^{2} = {(R-r)}^{2}$$
We still make use of the broadcast function of python.So self.sigma=[-sigma_big,sigma_small,sigma_small,...](one -sigma_big for one big particle,N sigma_smalls for N small particles ),using -sigma_big can solve this problem.But we notice that for the collisions between the big particle and small particles,it needs that $b^{2} - 4ac > 0$(small particles must be indide the big particle) and $c<0$,and we need the solution $-b + \frac{\sqrt{b^{2} - 4ac}}{2a}$.In the programming, we separated the two types of collisions in pair(big particle with small ones, small ones with small ones). And in this case, we just need to calculate the wall_time of the big particle.
